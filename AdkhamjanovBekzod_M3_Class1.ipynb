````python
# === SETUP — run this first ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')

url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
print('Loaded:', df.shape)
# Downloading data and changing TotalCharges values to numeric
````

### Cell 1 — see how many missing values per column

````python
df.isnull().sum()
````

### Cell 2 — check how many duplicate rows we have

````python
print('Duplicates:', df.duplicated().sum())
#Number of duplicates
````

### Cell 3 — make a copy and add a fake missing value to test fillna

````python
df_test = df.copy()
df_test.loc[0:9, 'tenure'] = np.nan
print('Missing tenure now:', df_test['tenure'].isnull().sum())
# Making a copy and replpacing first 10 values in tenure with NaN
````

### Cell 4 — fill missing tenure with the median

````python
median_tenure = df_test['tenure'].median()
df_test['tenure'] = df_test['tenure'].fillna(median_tenure)
print('After fillna, missing:', df_test['tenure'].isnull().sum())
print('Median used:', median_tenure)
# Calculating median for tenure and then filling NaN with it 
````

### Cell 5 — fill missing TotalCharges with 0

````python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
print('TotalCharges missing after fillna:', df['TotalCharges'].isnull().sum())
# Fill Totalcharges NaN values with 0 
````

### Cell 6 — drop duplicate rows

````python
before = len(df)
df = df.drop_duplicates()
after = len(df)
print(f'Before: {before}, After: {after}, Dropped: {before - after}')
# Before and after droping duplicate rows
````

### Cell 7 — convert SeniorCitizen from 0/1 to 'No'/'Yes'

````python
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
df['SeniorCitizen'].value_counts()
# Converting 0/1 values to booleans 
````

### Cell 8 — strip whitespace from gender column

````python
df['gender'] = df['gender'].str.strip()
df['gender'].value_counts()
# Stripping away blank spaces from strings
````

````python
import pandas as pd
import numpy as np 

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
print(df.shape)
print(df.head())
# Download file and print out shape and first 5 rows
````

````python
df.isnull.sum()
# Count missing values
````

````python
median_value = df["Age"].median()
df["Age"] = df['Age'].fillna(median_value)
print(df['Age'].isnull().sum())
# Calculating median and filling NaN with the value
````

````python
mode_embarked = df["Embarked"].mode[0]
df["Embarked"] = df['Embarked'].fillna(mode_embarked)
print('Mode used:', mode_embarked)
print('Missing Embarked:', df['Embarked'].isnull().sum())
# Find the most common value and replacing NaN with it
````

````python
df.drop(columns=['Cabin'], inplace=True)
print('Columns now:', df.shape[1])
# Dropping irrelecant column
````

````python
df.duplicated().sum()
````

````python
df["Survived_label"] = df["Survived"].map({0:"No", 1:"Yes"})
df[['Survived', 'Survived_label']].head()
````

````python
df['Sex'] = df["Sex"].str.strip().str.lower()
df["Sex"].value_counts()
# Cleaning string in Sex from blank spaces
````
