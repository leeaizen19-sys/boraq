## ▶️ Run this first — load the game

```python
import numpy as np
vocab = ['i', 'love', 'hate', 'pizza', 'you']

def to_vector(sentence, v=vocab):
    arr = np.zeros(len(v), dtype='float32')
    for w in sentence.lower().split():
        if w in v: arr[v.index(w)] = 1
    return arr

def relu(x): return max(0, x)

def brain(sentence, weights, v=vocab):
    vec = to_vector(sentence, v)
    score = float(np.sum(vec * weights))
    return vec, score, relu(score)

# ---- instant-feedback checkers (the 'game') ----
def check_vector(sentence, answer):
    real = [int(x) for x in to_vector(sentence)]
    you  = [int(x) for x in answer]
    if you == real:
        print(f'✅ CORRECT! "{sentence}" = {real}  🎉')
    else:
        print(f'❌ Not yet. You said {you}, but it should be {real}.')
        print('   Tip: put 1 where the word IS in the list, 0 where it is NOT.')

def check_guess(sentence, weights, guess):
    _, score, _ = brain(sentence, weights)
    if abs(guess - score) < 0.01:
        print(f'✅ SPOT ON! "{sentence}" scores {score}.  🎯')
    else:
        print(f'❌ Close? You guessed {guess}, real score is {score}.')
        print('   Tip: add up the weights of the words that ARE in the sentence.')

def test_brain(weights, rules):
    passed = 0
    for sentence, want in rules:
        _, _, out = brain(sentence, weights)
        got = 'happy' if out > 0 else 'sad'
        ok = (got == want)
        passed += ok
        print(f"{'✅' if ok else '❌'} {sentence:16s} -> {got.upper():5s} (need {want.upper()})")
    if passed == len(rules):
        print(f'\n🏆 ALL {len(rules)} PASSED — LEVEL CLEARED! 🎉')
    else:
        print(f'\n⏳ {passed}/{len(rules)} passed. Tweak the weights and run again!')

print('🎮 Game loaded. Good luck!')
```
