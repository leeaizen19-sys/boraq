```python
# ---------------------------------------------------------
# TASK 2: Handling Missing Values
# ---------------------------------------------------------
# 1. Fill missing numerical values with the median (middle value).
# This is better than the mean because it is not affected by outliers.
# Count missing values in each column
missing = df.isnull().sum()
# Calculate percentage of missing values
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

# Combine into a report DataFrame
missing_report = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0]
```

```python
# Select numerical columns (int, float)
numerical_cols = df.select_dtypes(include=[np.number]).columns
# Fill missing numerical values with median
# Median is robust to outliers
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing values with median: {median_val}")

# Select categorical columns (object/string).
categorical_cols = df.select_dtypes(include=['object']).columns
# Fill missing categorical values with mode (most frequent value)
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_val}")

# Verify that no missing values remain
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")
```

```python
# Count duplicate rows
n_duplicates = df.duplicated().sum()
print(f"Duplicate rows found: {n_duplicates}")

# Remove duplicates if any exist
if n_duplicates > 0:
    df = df.drop_duplicates()
    print(f"After removal: {df.shape[0]} rows remain")
else:
    print("No duplicates to remove.")
```

```python
# Check current data types of date-related columns
print("Before conversion:")
for col in df.columns:
    if 'date' in col.lower() or 'Date' in col:
        print(f"  {col}: {df[col].dtype}")
        print(f"  Sample value: {df[col].iloc[0]}")
```

```python
# Convert date columns from string to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Verify the conversion by checking the data types.
print("\nDate columns converted. New types:")
print(df[['Order Date', 'Ship Date']].dtypes)


```

```python
# First, identify the right column names
print("All columns:")
for col in df.columns:
    print(f"  {col}")
```

```python
# Display all column names to identify correct fields
for col in df.columns:
    print(col)

# Total spending per customer (sum of Sales)
customer_spending = df.groupby('Customer ID')['Sales'].sum()
customer_spending.name = 'total_spending'

# Order frequency per customer (number of orders)
order_freq = df.groupby('Customer ID').size()
order_freq.name = 'order_frequency'

# Average order value per customer
avg_order = df.groupby('Customer ID')['Sales'].mean()
avg_order.name = 'avg_order_value'

# Combine all features into a single DataFrame
customer_summary = pd.concat(
    [customer_spending, order_freq, avg_order],
    axis=1
)

# Display first 10 rows
customer_summary.head(10)
```

```python
# Save cleaned dataset
df.to_csv('superstore_cleaned.csv', index=False)

# Save customer summary
customer_summary.to_csv('customer_summary.csv')

print("Files saved successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv('superstore_cleaned.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])

```

```python
# Sales Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=50, color='steelblue', edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Profit Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Profit'], color='lightcoral')
plt.title('Boxplot of Profit')
plt.xlabel('Profit ($)')
plt.show()
```

**Interpretation:**
The Sales histogram is heavily right-skewed, meaning most transactions are for small amounts. In the Profit boxplot, the majority of the data is centered around zero, but there are significant outliers on both the positive (high profit) and negative (high loss) sides.

```python
# Category Sales
cat_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
cat_sales.plot(kind='bar', color=['#2196F3', '#4CAF50', '#FF9800'])
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.show()
```

```python

# Region Sales
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
region_sales.plot(kind='bar', color='steelblue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.show()
```

**Interpretation:**
Technology is the top-performing category in terms of revenue, while the West region generates the highest total sales. This suggests that the company's core market strength lies in the West and in high-value tech products.

```python
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Profit'] < lower_bound) | (df['Profit'] > upper_bound)]
print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
print(f"Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")
print(f"Number of outliers: {len(outliers)} out of {len(df)} rows")
```

**Interpretation:**
A high number of outliers were found, representing transactions that resulted in either exceptional profits or significant losses. These represent non-typical business cases that require individual attention.

```python
numerical_cols = df[['Sales', 'Quantity', 'Discount', 'Profit']]
corr_matrix = numerical_cols.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Heatmap: Sales, Quantity, Discount, Profit')
plt.show()
```

**Interpretation:**
Sales and Profit show a positive correlation, while Discount and Profit show a negative correlation. This confirms that higher discounts generally lead to lower profit margins per transaction.

```python
monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(color='steelblue', linewidth=2)
plt.title('Monthly Total Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid(True, alpha=0.3)
plt.show()
```

**Interpretation:**
The sales data shows a strong upward spike at the end of each year (November/December), indicating seasonality due to holiday shopping. There are also periodic dips at the beginning of each year.

**Why use `&` instead of `and`?**
We use `&` because it performs an "element-wise" comparison across the entire Series. Standard Python `and` is for single boolean values. Parentheses are required because bitwise operators have higher precedence than comparison operators.

In a `left` join, all rows from the original (left) table are kept. If there is no match in the lookup table, the new columns will be filled with `NaN` values for those rows.

## ðï¸ Section 1 â Exploratory vs Explanatory

**Task 1.1:**
*   **Exploratory visualization** is used during the data analysis phase to find patterns, trends, or outliers. It is "data for the analyst."
*   **Explanatory visualization** is used to communicate a specific insight to an audience. It is "data for the viewer."
*   In this assignment, I am producing **exploratory** visualizations as I am discovering the dataset's characteristics.

**Task 1.2:**
The lecture mentions **Anscombeâs Quartet** to demonstrate that four datasets can have identical summary statistics (mean, variance, correlation) yet look completely different when plotted. It proves that we must visualize data because summary statistics alone can be misleading.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure your dataset is uploaded to Colab
df = pd.read_csv('superstore_cleaned.csv')

# Matplotlib Histogram
plt.figure(figsize=(8, 4))
plt.hist(df["Sales"], bins=20, color='skyblue', edgecolor='black')
plt.title("Plain Matplotlib Histogram of Sales")
plt.xlabel("Sales ($)")
plt.ylabel("Count")
plt.show()

```

```python
# Matplotlib Scatter
plt.figure(figsize=(8, 4))
plt.scatter(df["Sales"], df["Profit"], alpha=0.5, color='green')
plt.title("Plain Matplotlib Scatter: Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
# Seaborn Histplot with KDE
plt.figure(figsize=(8, 4))
sns.histplot(df, x="Quantity", kde=True, color="orange")
plt.title("Seaborn: Quantity Distribution with KDE")
plt.show()

```

```python
# Seaborn Boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(df, x="Category", y="Profit", palette="Set2")
plt.title("Seaborn: Profit Distribution by Category")
plt.show()
```

```python
# Seaborn Scatterplot with Hue
plt.figure(figsize=(10, 6))
sns.scatterplot(df, x="Sales", y="Profit", hue="Region", style="Region")
plt.title("Seaborn: Sales vs Profit grouped by Region")
plt.show()
```

```python

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", center=0)
plt.title("Seaborn: Correlation Heatmap")
plt.show()
```

```python
# Using the Object-Oriented Pattern
fig, ax = plt.subplots(figsize=(8, 5))

# Plotting Ship Mode frequency
ship_counts = df['Ship Mode'].value_counts()
ax.bar(ship_counts.index, ship_counts.values, color='teal')

ax.set_xlabel("Ship Mode")
ax.set_ylabel("Number of Orders")
ax.set_title("Frequency of Shipping Modes")

# Saving the figure
fig.savefig("out.png", dpi=150)
print("File saved as out.png")
plt.show()
```

```python
# Visualizing heavy right-tail data
plt.figure(figsize=(12, 5))

# 1. Linear Scale
plt.subplot(1, 2, 1)
plt.hist(df["Sales"], bins=50, color='blue')
plt.title("Sales: Linear y-axis")

# 2. Log Scale
plt.subplot(1, 2, 2)
plt.hist(df["Sales"], bins=50, color='red')
plt.yscale("log")
plt.title("Sales: Logarithmic y-axis")

plt.show()
```

**Observation:**
In the linear histogram, the high-value sales were nearly invisible because the low-value sales were so frequent. Switching to a **log scale** revealed the distribution of those rare, high-value "long-tail" transactions that were previously hidden.
