```python
import pandas as pd

# Create a DataFrame (like a table)
# Keys are column names, values are data
df = pd.DataFrame({
    "name": ["Ali", "Zara", "Jasur"],     # names
    "age":  [22, 28, 31],                 # ages
    "city": ["Tashkent", "Samarkand", "Bukhara"]  # cities
})

df  # show the table
```

```python
# Select one column → returns Series (1D)
print(type(df['age']))

# Select multiple columns → returns DataFrame (2D)
print(type(df[['name', 'age']]))

# Index = row labels (default: 0,1,2,...)
print(df.index)
```

```python
df = pd.DataFrame({
    "name": ["Ali", "Zara", "Jasur"],
    "age":  [22, 28, 31],
    "city": ["Tashkent", "Samarkand", "Bukhara"]
})

print('--- before set_index ---')

# loc → select by label (here index is 0,1,2)
print(df.loc[0])

# iloc → select by position (0 = first row)
print(df.iloc[0])
```

```python
# Set 'name' as index
df = df.set_index('name')

print('--- after set_index ---')

# loc uses label (name)
print(df.loc['Ali'])

# iloc still uses position
print(df.iloc[0])

# This will give error because 0 is not index anymore
# print(df.loc[0])
```

```python
df = pd.DataFrame({
    "name": ["Ali", "Zara", "Jasur", "Madina", "Otabek"],
    "age":  [22, 28, 31, 17, 45],
    "city": ["Tashkent", "Samarkand", "Bukhara", "Tashkent", "Samarkand"]
})

# Correct filtering:
# & = AND, | = OR
# Use () for each condition
df[(df.age >= 18) & (df.city == 'Tashkent')]
```

```python
# People older than 25 and not in Tashkent
df[(df.age > 25) & (df.city != 'Tashkent')]
# People in Tashkent OR Bukhara
df[(df.city == 'Tashkent') | (df.city == 'Bukhara')]
# Youngest person
df[df.age == df.age.min()]
```

```python
# Group by city and calculate average age
df.groupby('city')['age'].mean()
```

```python
# Check result type
result = df.groupby('city')['age'].mean()

print(type(result))  # it is a Series
print(result)
```

```python
# Do multiple calculations at once
df.groupby('city').agg({
    'age': ['mean', 'min'],   # average and minimum age
    'name': 'count'           # number of people
})
```

```python
# Customers table
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Ali', 'Zara', 'Jasur', 'Madina', 'Otabek']
})

# Orders table
orders = pd.DataFrame({
    'customer_id': [1, 1, 3],
    'amount': [50000, 75000, 30000]
})

# LEFT JOIN → keep all customers
# If no order → NaN
left = pd.merge(customers, orders, on='customer_id', how='left')

# INNER JOIN → only matched rows
inner = pd.merge(customers, orders, on='customer_id', how='inner')

print('--- LEFT join ---')
print(left)

print('--- INNER join ---')
print(inner)
```

```python
# Region table
regions = pd.DataFrame({
    'city': ['Tashkent', 'Samarkand', 'Bukhara'],
    'region': ['Capital', 'South', 'South']
})

# Add city to customers
customers['city'] = ['Tashkent', 'Samarkand', 'Bukhara', 'Tashkent', 'Samarkand']

# Merge customers with region
cust_region = pd.merge(customers, regions, on='city', how='left')

# Merge with orders
full = pd.merge(cust_region, orders, on='customer_id', how='left')

# Group by region and sum amount
result = full.groupby('region')['amount'].sum()

# Show regions with total > 80000
result[result > 80000]
```
