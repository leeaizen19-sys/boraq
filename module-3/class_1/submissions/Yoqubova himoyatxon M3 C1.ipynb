# Module 3 | Class 1 Assignment — Data Cleaning Lab

```
import pandas as pd
import numpy as np
```

## Load Dataset

```
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

## Task 1 — Load and Inspect Dataset

```
print("Dataset Shape:")
print(df.shape)

print("\nFirst 5 rows:")
display(df.head())

print("\nData Types:")
print(df.dtypes)

print("\nDataset Info:")
df.info()

print("\nNumerical Summary:")
display(df.describe())

print("\nCategorical Summary:")
display(df.describe(include='object'))
```

```
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(numerical_cols)

print("\nCategorical Columns:")
print(categorical_cols)
```

## Task 2 — Missing Values

```
print(df.isnull().sum())
```

```
df[df['TotalCharges'] == ' ']
```

```
df['TotalCharges'] = pd.to_numeric(
    df['TotalCharges'],
    errors='coerce'
)

print(df.isnull().sum())
```

```
df['TotalCharges_missing'] = (
    df['TotalCharges']
    .isnull()
    .astype(int)
)

median_total = df['TotalCharges'].median()

df['TotalCharges'].fillna(
    median_total,
    inplace=True
)

for col in categorical_cols:
    
    if df[col].isnull().sum() > 0:
        
        mode_value = df[col].mode()[0]
        
        df[col].fillna(
            mode_value,
            inplace=True
        )

print(df.isnull().sum())
```

## Task 3 — Standardize Categorical Entries

```
for col in categorical_cols:
    
    print("\nColumn:", col)
    
    print(df[col].value_counts())
```

```
for col in categorical_cols:
    
    df[col] = df[col].str.strip()
    df[col] = df[col].str.lower()
```

```
for col in categorical_cols:
    
    print("\nCleaned Column:", col)
    
    print(df[col].value_counts())
```

## Task 4 — Data Types and Feature Engineering

```
print(df['TotalCharges'].dtype)

df['tenure_years'] = df['tenure'] / 12

print(df.dtypes)
```

## Task 5 — Detect and Remove Duplicates

```
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)

display(df[df.duplicated(keep=False)])
```

```
df = df.drop_duplicates()

print(df.shape)
```

## Task 6 — Identify Unrealistic Values

```
print("Negative Monthly Charges:")
display(df[df['MonthlyCharges'] < 0])

print("\nTenure Statistics:")
display(df['tenure'].describe())

print("\nCustomers with tenure > 72:")
display(df[df['tenure'] > 72])

print("\nCustomers with tenure = 0:")
display(df[df['tenure'] == 0][['tenure', 'TotalCharges']])
```

## Final Dataset Information

```
df.info()
```

## Save Cleaned Dataset

```
df.to_csv(
    'telco_cleaned.csv',
    index=False
)

print("Cleaned dataset saved successfully!")
```

# Data Cleaning Summary

In this project, the Telco Customer Churn dataset was cleaned and prepared for analysis.

First, the dataset structure, data types, and statistical summaries were inspected. During the cleaning process, missing values were identified, especially in the TotalCharges column where blank spaces were stored as strings. These values were converted into NaN using pd.to_numeric() with errors='coerce'.

A missing indicator column called TotalCharges_missing was created to preserve information about missingness. Numerical missing values were filled using the median because the median is resistant to outliers. Categorical missing values were filled using the mode.

Categorical columns were standardized by removing extra spaces and converting text into lowercase to avoid inconsistent categories such as “Yes”, “YES”, and “yes”.

Duplicate rows were checked and removed if necessary. Numerical columns were also inspected for unrealistic values such as negative charges or invalid tenure values.

Finally, a new feature called tenure_years was created from the tenure column, and the cleaned dataset was saved as telco_cleaned.csv.
