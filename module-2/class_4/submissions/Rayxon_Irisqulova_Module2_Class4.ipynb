**Task 1: Distribution Plots**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('superstore_cleaned.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=50, color='steelblue', edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.show()
```

**The Sales distribution is highly right-skewed. As seen in the histogram, the vast majority of transactions (over 8,000) are concentrated in the $0 to $1,000 range. This indicates that the business relies heavily on high-volume, low-value transactions. The "long tail" extending towards $20,000 shows that while expensive sales occur, they are rare exceptions rather than the norm.**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Profit'], color='lightcoral')
plt.title('Boxplot of Profit')
plt.xlabel('Profit ($)')
plt.show()
```

In the Profit boxplot, the quartiles (the central box) are extremely compressed around the zero mark. This suggests that for 50% of the data, the profit or loss per transaction is very small. However, there are numerous visible outliers on both ends of the spectrum:
Negative Outliers: Some transactions resulted in significant losses, reaching as low as -$6,000.

Positive Outliers: On the other hand, certain "home run" sales generated profits as high as +$8,000

**Task 3: Outlier Identification**

```python
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")
```

```python
outliers = df[(df['Profit'] < lower_bound) | (df['Profit'] > upper_bound)]

print(f"Number of outliers: {len(outliers)} out of {len(df)} rows ({len(outliers)/len(df)*100:.2f}%)")
```

**I identified 1,881 outliers, which account for 18.82% of the total dataset. Based on the previous Profit boxplot, these outliers consist of both high-profit and high-loss items. While there are extreme losses reaching below -$6,000, there is a slightly more visible concentration of high-profit outliers extending up to approximately +$8,000. This indicates that while the business experiences significant volatility, it also benefits from occasional high-revenue transactions that deviate from the normal profit range.**

**The chart reveals a clear seasonal pattern where sales consistently increase towards the end of each year. Significant spikes occur every November and December, likely due to holiday shopping seasons. Overall, the business shows a positive year-over-year growth trend despite these seasonal fluctuations.**

**Task 6: Reflection Questions**

**1. Why is retail sales data often right-skewed?**
In retail, the vast majority of transactions consist of everyday, low-to-medium value items, which creates a large "peak" on the left side of the histogram. However, there are occasionally very high-value sales—such as bulk corporate orders or luxury electronics—that create a "long tail" extending to the right. This imbalance between frequent small purchases and rare large ones results in a right-skewed distribution.

**2. Should you remove outliers in Profit? What information might you lose?**
No, profit outliers should not be removed because they represent the most significant financial events, such as "home run" sales or critical losses. By deleting them, you lose vital information about the business's volatility and its exposure to extreme financial risks or opportunities. These outliers often provide the most important insights into which specific strategies are working or failing.

**3. If Discount and Profit have a negative correlation, does that mean the company should stop giving discounts?**
Not necessarily, because correlation does not equal causation. While higher discounts are associated with lower profit margins per item, they may be a strategic necessity to increase total sales volume, clear out old inventory, or acquire new customers. The goal should be to find an "optimal" discount level rather than eliminating them entirely.

**4. As a store manager, what action would you take based on the monthly sales trend?**
Based on the clear seasonal spikes in November and December, I would increase inventory levels and hire seasonal staff in advance to handle the holiday peak. To address the notable dips in January and February, I would plan aggressive marketing campaigns or "New Year" clearance sales to stimulate demand during the post-holiday slowdown.
