```
import pandas as pd

# Creating the DataFrame
df = pd.DataFrame({
    "name": ["Ali", "Zara", "Jasur"],
    "age": [22, 28, 31],
    "city": ["Tashkent", "Samarkand", "Bukhara"]
})

# Answers to the questions:
print(type(df["age"]))           # Returns: pandas.core.series.Series
print(type(df[["name", "age"]])) # Returns: pandas.core.frame.DataFrame
print(df.index)                  # Returns: RangeIndex(start=0, stop=3, step=1).
                                 # By default, it is a numeric range starting from 0.
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset from Class 4
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

### Task 2.1 — Plain Matplotlib
plt.figure(figsize=(8, 4))
plt.hist(df["Age"].dropna(), bins=20)
plt.title("Distribution of Age")
plt.show()

plt.figure(figsize=(8, 4))
plt.scatter(df["Age"], df["Fare"])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

### Task 2.2 — Seaborn
# Histogram with KDE
sns.histplot(df, x="Age", kde=True)
plt.show()

# Boxplot
sns.boxplot(df, x="Pclass", y="Age")
plt.show()

# Scatterplot with Hue
sns.scatterplot(df, x="Age", y="Fare", hue="Survived")
plt.show()

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()

### Task 2.3 — OO Matplotlib for control
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_title("Squares")
fig.savefig("out.png", dpi=150)
```

```
import pandas as pd
import numpy as np

# Step 1-2: Load dataset
# Eslatma: Kaggle'dan yuklangan fayl nomi 'WA_Fn-UseC_-Telco-Customer-Churn.csv' bo'lishi kerak
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Step 3: Shape
print(f"Dataset shape: {df.shape}")

# Step 4: Types and Info
print(df.info())

# Step 5: Numerical and Categorical summary
print(df.describe())
print(df.describe(include='object'))

# Step 6: Identify types
# Numerical: tenure, MonthlyCharges, TotalCharges (currently object)
# Categorical: gender, SeniorCitizen, Partner, Dependents, etc.
```

```
# Step 1: Count NaN
print(df.isnull().sum())

# Step 2: Hidden blanks in TotalCharges
print(f"Blank strings: {(df['TotalCharges'] == ' ').sum()}")

# Step 3: Convert to numeric (coercing errors creates NaNs)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Step 4: Classification (Note)
# Missing 'TotalCharges' usually happens for new customers (tenure=0).
# This is MAR (Missing At Random).

# Step 5: Missing indicator
df['TotalCharges_missing'] = df['TotalCharges'].isnull().astype(int)

# Step 6: Impute numerical (Median)
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

# Step 7: Impute categorical (Mode)
# Agar categorical ustunlarda missing bo'lsa:
# df['col'].fillna(df['col'].mode()[0], inplace=True)
```

```
# Step 1-2: Inspect categorical counts
# Masalan: df['MultipleLines'].value_counts()

# Step 3: Strip whitespace
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Step 4: Standardize case/values
# Misol: Ba'zi joylarda 'No phone service' ni 'No' ga o'zgartirish kerak bo'lsa map ishlatiladi
```

```
# Step 1: Confirm TotalCharges is float64
print(df['TotalCharges'].dtype)

# Step 2: Check dates (In this dataset, there are no specific date columns)
# If existed: df['Date'] = pd.to_datetime(df['Date'])

# Step 3: Temporal feature
df['tenure_years'] = df['tenure'] / 12

# Step 4: Final type check
print(df.dtypes)
```

```
# Step 1: Count duplicates
print(f"Duplicates: {df.duplicated().sum()}")

# Step 2-3: Remove duplicates
df = df.drop_duplicates()

# Step 4: Document
# Agar duplicates bo'lsa, nechta o'chirilganini print qiling
```
