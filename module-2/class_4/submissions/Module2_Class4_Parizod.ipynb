### 3a. Selection — `loc` vs `iloc`

**Concept.** `loc` selects by **label**. `iloc` selects by **integer position**. They behave the same only when your index happens to be `0, 1, 2, …` — once you set a meaningful index, they diverge.

### 3b. Boolean Filtering
We use & because it performs an element-wise comparison on the entire Series, whereas the Python and keyword is designed for single boolean values. Parentheses are required because the bitwise operators (&, |) have higher precedence than comparison operators.

# C

**Correct one:** ?  
**Why each wrong one fails:**

C fails because: It uses the && operator, which is common in languages like C++ or Java but is not a valid operator in Python or Pandas.

D fails because: The df.filter() method is used for filtering labels (row/column names), not for checking the actual data values inside the rows like age == 18.

Use `.agg()` with a dict to compute, in **one** call:
- average `age` per city

### 3d. Merging

### 🤔 Predict

I have two DataFrames. Customers (5 rows) and Orders (3 rows, only some customers have ordered). I run:

```python
pd.merge(customers, orders, on='customer_id', how='left')
```

Predict:
- How many rows in the result?
- For customers without orders, what fills the order columns?
- If I changed `how='inner'`, how many rows would I get?

````python
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Ali', 'Zara', 'Jasur', 'Madina', 'Otabek']
})
orders = pd.DataFrame({
    'customer_id': [1, 1, 3],
    'amount': [50000, 75000, 30000]
})

left = pd.merge(customers, orders, on='customer_id', how='left')
inner = pd.merge(customers, orders, on='customer_id', how='inner')

print('--- LEFT join ---')
print(left)
print()
print('--- INNER join ---')
print(inner)
````

### ✍️ Explain

1. Why does the LEFT join have 6 rows but the INNER has only 3?
The LEFT join preserves all rows from the original customers table, including those who haven't placed an order, whereas the INNER join only keeps rows where a match exists in both tables.
2. Notice that customer 1 appears twice in LEFT — explain why.
Customer 1 (Ali) appears twice because he has two corresponding records in the orders table; the join creates a new row for every match found.
3. *Quick Check from the lecture:* In one sentence, what is the rule for what happens to unmatched left rows in a LEFT join?
Unmatched rows from the left table are kept in the final result, and the missing data from the right table is filled with NaN (null) values.

### 🛠 Practice — capstone for today

Combine everything in this section:
1. Add a `region` lookup DataFrame (Tashkent → 'Capital', Samarkand → 'South', Bukhara → 'South', etc.).
2. Merge it into your customer DataFrame.
3. GroupBy region and compute total order `amount` per region (combine with the orders DataFrame).
4. Filter to show only regions with total > 80000.
5. In a markdown cell, write 2–3 sentences explaining the business interpretation of your result.

````python
# Capstone work below
# 1. Add a region lookup DataFrame
regions = pd.DataFrame({
    'city': ['Tashkent', 'Samarkand', 'Bukhara'],
    'region': ['Capital', 'South', 'South']
})

# 2. Merge region into customer DataFrame
df_with_regions = pd.merge(customers, df, on='name', how='left') # assuming 'df' from section 3b has cities
df_with_regions = pd.merge(df_with_regions, regions, on='city', how='left')

# 3. Combine with orders and GroupBy region for total amount
final_df = pd.merge(df_with_regions, orders, on='customer_id', how='left')
region_totals = final_df.groupby('region')['amount'].sum().reset_index()

# 4. Filter to show only regions with total > 80000
high_value_regions = region_totals[region_totals.amount > 80000]
print(high_value_regions)
````
